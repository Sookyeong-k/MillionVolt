### 전처리 2 : 결측치 처리
- 요일 : 평일 / 주말
- 시간별 집계치 : 집계오류 등으로 인한 집계 데이터가 없음 -> 평균으로처리

```python
traffic_df['요일(2)'].unique()
```

```python
# 평일, 주말 채우기
traffic_df['요일(2)'] = traffic_df.apply(lambda x : '주말' if x['요일'] in ['토', '일'] else '평일', axis = 1)
#데이터 확인
traffic_df['요일(2)'].unique()
```

```python
#컬럼확인
print(traffic_df.columns)

#교통량 결측치 채우기 : 각 지점번호의 평균치 -> 데이터프레임 생성
traffic_point_mean_df = pd.DataFrame(traffic_df.groupby('지점번호')['0시', '1시', '2시', '3시', '4시',
       '5시', '6시', '7시', '8시', '9시', '10시', '11시', '12시', '13시', '14시', '15시',
       '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시'].mean().round(2))
```

```python
#결측치를 traffic_point_mean_df의 지점번호 시간대 평균값으로 채우기

# traffic_columns 정의
traffic_columns = ['일자', '요일', '지점명', '지점번호', '방향', '구분', '0시', '1시', '2시', '3시', '4시',
       '5시', '6시', '7시', '8시', '9시', '10시', '11시', '12시', '13시', '14시', '15시',
       '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시', '요일(2)']

# traffic_df의 열 이름을 traffic_columns으로 설정
traffic_df.columns = traffic_columns

# 결측치를 채우는 함수 정의
def fill_traffic_na(row):
    for col in traffic_columns:
        if pd.isnull(row[col]):
            point_num = row['지점번호']
            if point_num in traffic_point_mean_df.index:
                mean_value = traffic_point_mean_df.loc[point_num, col]
                row[col] = mean_value
    return row

# 결측치를 채운 데이터프레임 얻기
traffic_df = traffic_df.apply(fill_traffic_na, axis=1)
```

- 좌표로 행정동명 삽입(카카오API)

```python
#데이터프레임 복사
traffic_point_hdong_df = traffic_point_df.copy()
```

```python
import requests
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

def get_admin_district_name(latitude, longitude, api_key):
    url = f'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x={longitude}&y={latitude}'
    headers = {'Authorization': f'KakaoAK {api_key}'}
    response = requests.get(url, headers=headers)
    data = response.json()
    if data.get('documents'):
        region = data['documents'][0]
        region_name = region['region_3depth_name']
        return region_name
    else:
        return None

# 카카오 API 키
api_key = '13ee922882dc101cbce7c7a616661d77'

# 데이터프레임에서 위도와 경도 정보 추출
coordinates = [(row['위도'], row['경도']) for index, row in traffic_point_hdong_df.iterrows()]

# 중복된 좌표 제거
unique_coordinates = list(set(coordinates))

# API 요청 결과 캐싱
cached_results = {}
for coord in unique_coordinates:
    cached_results[coord] = None

# 캐싱되지 않은 결과에 대해서만 API 요청
def process_coordinate(coord):
    latitude, longitude = coord
    if cached_results[coord] is None:
        cached_results[coord] = get_admin_district_name(latitude, longitude, api_key)

# 병렬 처리를 사용하여 API 요청 동시에 처리
with ThreadPoolExecutor() as executor:
    executor.map(process_coordinate, unique_coordinates)

# 결과를 DataFrame에 추가
traffic_point_hdong_df['행정동명'] = [cached_results[coord] for coord in coordinates]

```
